**Q4**-Why LookupError class is used? Explain with an example KeyError and IndexError.

**Answer**:The LookupError class is a built-in exception class in Python that serves as the base class for a number of specific lookup-related exception classes. It is used when an index or key cannot be found in a sequence or mapping, respectively.

Two common examples of specific lookup-related exception classes that inherit from LookupError are KeyError and IndexError.

```python
## example for keyerror:
di={"a":1,"b":2,"c":3}
value=di["d"]
```

**Q5**:Explain ImportError. What is ModuleNotFoundError?

**Answer**:when we import a module that is not found or has an error, we will receive an ImportError. This means that Python cannot locate or import the module that we specified in your code.Reason of import error :

(1) The module is not installed in our system .

(2) The module is not on the search path of our puthon enviourment.

(3) The module has syntax errors or other issues that prevent it from being loaded properly.

**ModuleNotFoundError:** This exception is a subclass of ImportError, and it is raised when a module is not found in the search path.
